```python
import csv
import sys
csv.field_size_limit(sys.maxsize)
news_dataset=pd.read_csv('/content/WELFake_Dataset.csv', engine='python')
```

```python
news_dataset=news_dataset.fillna('')
```

```python
x=news_dataset.drop(columns='label',axis=1)
y=news_dataset['label']
```

```python
 port_stem=PorterStemmer()
```

```python
def stemming(text):
  stemmend_content=re.sub('[^a-zA-Z]',' ',text)
  stemmend_content=stemmend_content.lower()
  stemmend_content=stemmend_content.split()
  stemmend_content=[port_stem.stem(word) for word in stemmend_content if not word in stopwords.words('english')]
  stemmend_content=' '.join(stemmend_content)
  return stemmend_content
```

```python
from tqdm import tqdm
tqdm.pandas()

news_dataset['title'] = news_dataset['title'].progress_apply(stemming)
```

```python
print(news_dataset['title'])
```

```python
#separating the data and label
X = news_dataset['title'].values
Y = news_dataset['label'].values
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# converting the textual data to numerical data
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy score on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```
